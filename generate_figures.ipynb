```python
import os
import MacUtils as mut
from importlib import reload
from matplotlib import pyplot as plt

output_dir = "results/fig"
os.makedirs(output_dir, exist_ok=True)

panels = [
    ('a', 'colorectal_cancer', mut.figure4a),
    ('b', 'hcv_liver_disease', mut.figure4b),
    ('c', 'sepsis', mut.figure4c),
    ('d', 'idiopathic_pulmonary_fibrosis', mut.figure4d),
    ('e', 'kidney_transplant_event_free', mut.figure4e),
    ('f', 'kidney_transplant_overall', mut.figure4f),
    ('g', 'vitality_study', mut.figure4g),
]

for label, name, fn in panels:
    try:
        print(f"Generating panel {label} ({name})...")
        result = fn()

        # ---- CASE 1: multiple figures returned (like figure4g) ----
        if isinstance(result, tuple):
            for i, obj in enumerate(result):
                if hasattr(obj, 'get_figure'):
                    fig = obj.get_figure()
                else:
                    fig = plt.gcf()

                filepath = os.path.join(output_dir, f"{name}_part{i+1}.pdf")
                fig.savefig(filepath, bbox_inches='tight')
                plt.close(fig)

                print(f"  Saved: {filepath}")

        # ---- CASE 2: single figure ----
        else:
            if hasattr(result, 'get_figure'):
                fig = result.get_figure()
            else:
                fig = plt.gcf()

            filepath = os.path.join(output_dir, f"{name}.pdf")
            fig.savefig(filepath, bbox_inches='tight')
            plt.close(fig)

            print(f"  Saved: {filepath}")

    except Exception as e:
        print(f" Panel {label} — {e}")
        plt.close('all')

print("All panels saved .")
```
